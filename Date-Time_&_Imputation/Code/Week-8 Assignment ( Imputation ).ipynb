# Code regarding first data frame
# (xls file :- imputation_homework_04.xls)

```python
# Number of null values in the dataframe.
df1.isnull().sum().sum()
```

```python
df1.head()
```

```python
# Deleting the unwanted column from the dataframe.
del df1["Unnamed: 0"]
```

```python
# Number of missing values per each column in the dataframe. 
df1.isnull().sum()
```

```python
import matplotlib.pylab as plt
```

```python
# Histogram plot for of distribution of poins in values column
hist = df1.hist(column='value',bins=100,alpha = 1)
```

```python
# Lag plot of time column
from pandas.plotting import lag_plot
lag_plot(df1.time)
```

```python
# Lagplot of value column
lag_plot(df1.value)
```

From the above lag plot, I dont think that the order in the data matters

```python
# As the order in the data doesn't matter, I want to use the ffill() function to fill the missing values along the index axis.
# The missing value is filled based on the corresponding value in the previous row.
df1 = df1.fillna(method = "ffill")
```

```python
# Expanding the output display to see all the rows and columns in the data frame
# Printing the modified data frame
pandas.set_option("display.max_rows", None, "display.max_columns", None)
print(df1)
```

```python
# Scatter plot using columns in the dataframe.
plt.scatter(df1.time,df1.value)
```
