```python
from zipfile import ZipFile 
from bs4 import BeautifulSoup as bs 
import xml.etree.ElementTree as ET
```

```python
# Using zip keyword to open the zip file
with ZipFile("xml_containing_html.xml.zip", 'r') as zip:  
# Using extractall method to extract the files
    zip.extractall()
```

```python
# Getting data through element tree 
tree = ET.parse('xml_containing_html.xml') 
#reading the XMl file that was extravted before.
root = tree.getroot()
#Using beautiful soup for parsing
num=0 
for j in root:   
    for i in j:
        data=i.text
        soup=bs(data,'html.parser') 
        for link in soup.find_all('a'): 
            ab=(link.get('href'))
            print(ab)
            num+=1
            
    print('Number of links in this HTML page is  ',num)
   

```
