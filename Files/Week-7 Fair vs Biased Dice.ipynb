# week 7: Simulate a fair die and a biased 6-sided die


Algorithm of my code.

1) First, I am taking two lists . list1 consists of numbers 1 to 6 which can be used to simulate a fair dice. list2 consists of    three elements each from 1 to 5 and five 6's which can be used to simulate a baised dice.

2) When we select a random number number using random.choice function from list1 it acts similar to that of simulating a fair      dice.

3) When we select a random number using random.choice functionfrom list2, the probability of selecting numbers from 1 to 5 is      0.15 (3/20) and the probability of getting number 6 is 0.25 (5/20).

4) I defined a function named dice with an argument n.

5) In the dice function, I have randomly choosen n elements from list1 and n elements from list2 and appended those elements to    two new listsx and listy. list1 resembles the outcomes of fair dice and list2 resembles the outcomes of the biased dice. 

6) Now I have take the count of each element from 1 to 6 from the fair dice outcomes that is listx and biased dice outcoms that   is listy into to new lists lista and listb.

7) Using matplotlib, I have compared the outcomes of fair and biased dice with double bar graph.

```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
list1=[]
list2=[]
for i in range(1,7):
    list1.append(i)
for i in range(1,6):
    for j in range(0,3):
        list2.append(i)
for i in range(0,5):
    list2.append(6)

```

```python
def dice(n):
    listx=[]
    listy=[]
    lista=[]
    listb=[]
    for i in range(0,n):
        listx.append(random.choice(list1))
        listy.append(random.choice(list2))
    
    
    for i in range(1,7):
        lista.append(listx.count(i))
        listb.append(listy.count(i))
    x=np.arange(6)
    plt.figure(figsize=(20,10))
    plt.bar(x + 1.00,lista,color = 'b',width = 0.25,label='Fair Dice')
    plt.bar(x + 1.25,listb,color = 'g',width = 0.25,label='Biased Dice')
    plt.xlabel('Out come of  Dice',size=20)
    plt.ylabel('Number of times',size=20)
    plt.legend(['Fair Dice','Biased Dice'],prop={'size': 20})
    plt.show()
                     
```

```python
dice(50)
```
