```python
# Getting the mean values of demand of electricity from every hour in dframe1 (march1) and storing those values in new dataframe dframe11 
dframe11=dframe1.resample('h').mean()
```

```python
# The header of new data fame looks like
dframe11.head()
```

```python
# Storing the mean values present in the data frame into a list
list1 = dframe11['value1'].to_list()
```

```python
# Getting the mean values of demand of electricity from every hour in remaining dataframes and storing those values in new dataframes
dframe22=dframe2.resample('h').mean()
dframe33=dframe3.resample('h').mean()
dframe44=dframe4.resample('h').mean()
dframe55=dframe5.resample('h').mean()
dframe66=dframe6.resample('h').mean()
dframe77=dframe7.resample('h').mean()
dframe88=dframe8.resample('h').mean()
dframe99=dframe9.resample('h').mean()
dframe110=dframe10.resample('h').mean()



```

```python
# Storing the mean values present in the data frames into new lists
list2 = dframe22['value2'].to_list()
list3 = dframe33['value3'].to_list()
list4 = dframe44['value4'].to_list()
list5 = dframe55['value5'].to_list()
list6 = dframe66['value6'].to_list()
list7 = dframe77['value7'].to_list()
list8 = dframe88['value8'].to_list()
list9 = dframe99['value9'].to_list()
list10 = dframe110['value10'].to_list()
```

```python
# Creating a new dataframe
dframe= pd.DataFrame()
```

```python
# I the new dataframe created, every list created above is added as a column to the new list
dframe['value1'] = np.array(list1)
dframe['value2'] = np.array(list2)
dframe['value3'] = np.array(list3)
dframe['value4'] = np.array(list4)
dframe['value5'] = np.array(list5)
dframe['value6'] = np.array(list6)
dframe['value7'] = np.array(list7)
dframe['value8'] = np.array(list8)
dframe['value9'] = np.array(list9)
dframe['value10'] = np.array(list10)
```

```python
# Modified dataframe looks like
dframe
```

```python
# Calculating the average across all days for an hour
dframe['AVG'] = np.array(dframe.mean(axis=1))
```

```python
dframe.head()
```

```python
# Taking the averages values in a list
list_Avg = dframe['AVG'].to_list()
```

```python
print(list_Avg)
```

```python
import matplotlib
print('maplotlib',matplotlib.__version__)
import matplotlib.pyplot as plt
```

```python
# Plotting the average consumption of electricity across all days for an hour vs hours 
label=('00:00:00 - 1:00:00','1:00:00 - 2:00:00','2:00:00 - 3:00:00','3:00:00 - 4:00:00','4:00:00 - 5:00:00','5:00:00 - 6:00:00','6:00:00 - 7:00:00','7:00:00 - 8:00:00','8:00:00 - 9:00:00','9:00:00 - 10:00:00','10:00:00 - 11:00:00','11:00:00 - 12:00:00','12:00:00 - 13:00:00','13:00:00 - 14:00:00','14:00:00 - 15:00:00','15:00:00 - 16:00:00','16:00:00 - 17:00:00','17:00:00 - 18:00:00','18:00:00 - 19:00:00','19:00:00 - 20:00:00','20:00:00 - 21:00:00','21:00:00 - 22:00:00','22:00:00 - 23:00:00','23:00:00 - 00:00:00')
y_pos = np.arange(len(label))
f, sub = plt.subplots(figsize=(30,15))
plt.bar(y_pos, list_Avg, align='center', alpha=0.5,color= 'b')
plt.xticks(y_pos, label,rotation='vertical',size='20')
plt.yticks(size=20)
plt.xlabel('Time',size=30)
plt.ylabel('Demand',size=30)
plt.title('Average Usage of Electriciy for all days across an hour',size = 30)
plt.show()
```
